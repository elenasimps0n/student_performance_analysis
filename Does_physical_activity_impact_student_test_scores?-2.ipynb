This project aims to explore how the physical activity level of students impacts their test scores.




"Dataset: Student Performance Factors (6,607 students). Source: Kaggle. Contains variables including hours studied, attendance, sleep hours, physical activity, and exam scores." https://www.kaggle.com/datasets/grandmaster07/student-exam-performance-dataset-analysis?resource=download&select=StudentPerformanceFactors.csv


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
avg_scores = data.groupby('Physical_Activity')['Exam_Score'].mean()
x_axis = np.linspace(0,5,5)

avg_scores.plot(kind='line')
plt.xlabel('Physical Activity (hours per week)')
plt.ylabel('Average Exam Score')
plt.title('Average Exam Score by Physical Activity Level')

plt.show()
```

Students who do more physical activity per week tend to achieve higher average exam scores. The trend is broadly positive, with students doing 6 hours of activity scoring notably higher than those doing 1-2 hours. There is a dip at 1 hour before the scores rise, implying that the small amount of physical activity is not enough to observe a difference in test scores. As the hours of physical activity increase, there is a more significant increase in the scores. Overall, this shows a positive correlation between physical activity and academic performance, though other factors in the dataset (such as hours studied and attendance) likely also may have impacts on the scores.
